```python
comments_no_punctuations = df['textOriginal'].replace('.', '').replace('!', '').replace('!','').replace('?','')
lower_case = comments_no_punctuations.str.lower()
words = lower_case.str.split()


len(words)
words.head(10)
```

```python
# Replace any NaN values in the 'words' Series with empty lists
words_cleaned = words.fillna('').apply(list)

total_words = words_cleaned.apply(len).sum()
print(f"Total number of words: {total_words}")
```

```python
# Assuming 'words_cleaned' is the Series containing the lists of words

words_to_remove = ['the', 'to', 'and', 'is', 'a', 'of', 'it', 'in', 'i', 'you', 'that', 'for', 'on', 'are', 'with', 'not','have', 'just', 'as', 'your', 'my', 'was',
                   'this', 'they', 'be', 'but', 'can', 'will', 'so', 'if', 'from', 'like', 'what', 'about', 'all', 'or', 'we',
                   'how', 'their', "it's", 'by', 'do', 'has', 'at', 'an', 'its', 'no', 'get', 'when', 'would', 'than']

filtered_words = words_cleaned.apply(lambda x: [word for word in x if word not in words_to_remove])

# Now 'filtered_words' is your new Series with the specified words removed
print(filtered_words.head())
```

```python
# Define the list of keywords to filter by
keywords = ["AI", "Tech", "Model", "Models", "DeepSeek", "Openai", "use", "China", "US", "Chinese", "Open", "source"]

# Filter the DataFrame to keep comments that contain any of the keywords (case-insensitive)
filtered_df = df[df['textOriginal'].str.contains('|'.join(keywords), case=False, na=False)]

# Display the filtered comments
print("Comments containing the specified keywords:")
display(filtered_df[['textOriginal', 'videoId', 'channelId']])
```

```python
import re

# Define the list of keywords to filter by
keywords = ["AI", "Tech", "Model", "Models", "DeepSeek", "Openai", "use", "China", "US", "Chinese", "Open", "source"]

# Create a dictionary to store keyword counts
keyword_comment_counts = {}

# Count comments for each keyword
for keyword in keywords:
    if keyword == "US":
        # Use regex with word boundaries to match only the whole word "US" (case-insensitive)
        count = df[df['textOriginal'].str.contains(r'\bUS\b', case=False, na=False)].shape[0]
    else:
        count = df[df['textOriginal'].str.contains(keyword, case=False, na=False)].shape[0]
    keyword_comment_counts[keyword] = count

# Display the keyword comment counts in a table
keyword_counts_df = pd.DataFrame.from_dict(keyword_comment_counts, orient='index', columns=['Comment Count'])
print("Keyword occurrences in comments:")
display(keyword_counts_df)

# Calculate the total number of comments
total_comments = df.shape[0]

# Calculate the number of comments containing at least one keyword
# Need to re-calculate this based on the refined matching for "US"
comments_with_keywords = df[df['textOriginal'].str.contains('|'.join([r'\bUS\b' if k == 'US' else k for k in keywords]), case=False, na=False)].shape[0]


# Calculate the number of comments without any keywords
comments_without_keywords = total_comments - comments_with_keywords

print(f"\nTotal number of comments: {total_comments}")
print(f"Number of comments containing at least one keyword: {comments_with_keywords}")
print(f"Number of comments without any of the specified keywords: {comments_without_keywords}")
```
